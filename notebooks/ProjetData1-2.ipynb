```python
def generate_box_plot(data, attribute_name, show_outliers=True, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
    
    if show_outliers:
        ax.boxplot(data, vert=False, patch_artist=True, labels=[attribute_name])
    else:
        ax.boxplot(data, vert=True, patch_artist=True, labels=[attribute_name], showfliers=False)
    
    ax.set_title(f"BoÃ®te Ã  Moustache de {attribute_name}")
    ax.grid(True)
```

## Le traitement des valeurs aberrantes

```python
def calculer_quartiles(attribut):
    # Trier l'attribut
    attribut_trie = sorted(attribut)

    # Calculer (Q0) qui est le min
    Q0 = attribut_trie[0]

    # Calculer le premier quartile (Q1)
    n = len(attribut_trie)
    q1_idx = n // 4
    Q1 = (attribut_trie[q1_idx - 1] + attribut_trie[q1_idx]) / 2

    # Calculer le deuxiÃ¨me quartile (Q2, la mÃ©diane)
    if n % 2 == 1:
        Q2 = attribut_trie[n // 2]
    else:
        mid_idx = n // 2
        Q2 = (attribut_trie[mid_idx - 1] + attribut_trie[mid_idx]) / 2

    # Calculer le troisiÃ¨me quartile (Q3)
    q3_idx = (3 * n) // 4
    Q3 = (attribut_trie[q3_idx - 1] + attribut_trie[q3_idx]) / 2

    # Calculer le quatriÃ¨me quartile (Q4)
    q4_idx = n - q1_idx
    Q4 = (attribut_trie[q4_idx - 1] + attribut_trie[q4_idx]) / 2


    return Q0, Q1, Q2, Q3, Q4
```

```python
def remove_outliers_mean(dfMoy, column_name):
    # Extraction de la colonne spÃ©cifiÃ©e
    column_data = dfMoy[column_name]
    
    # VÃ©rification si la colonne est 'Start date' ou 'end date'
    if column_name not in ['Start date', 'end date']:
        # Calcul de la moyenne et de l'Ã©cart type de la colonne
        mean_value = column_data.mean()

        # Calcul des quartiles pour identifier les valeurs aberrantes
        q1 = column_data.quantile(0.25)
        q3 = column_data.quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        # Remplacement des valeurs aberrantes par la moyenne
        dfMoy[column_name] = np.where((column_data < lower_bound) | (column_data > upper_bound), mean_value, column_data)
```

```python
# Conversion des colonnes de dates en objets datetime
dfMoy['Start date'] = pd.to_datetime(dfMoy['Start date'], errors='coerce')
dfMoy['end date'] = pd.to_datetime(dfMoy['end date'], errors='coerce')

# Suppression des valeurs aberrantes et gÃ©nÃ©ration de boÃ®tes Ã  moustaches
for c in col:
    remove_outliers_mean(dfMoy, c)
```

### Quel est le rapport entre les cas confirmÃ©s, les tests effectuÃ©s et les tests positifs au fil du temps pour chaque zone ? ( La pÃ©riode doit Ãªtre choisie)

```python
from scipy.interpolate import make_interp_spline
import pandas as pd
import matplotlib.pyplot as plt

# Assume que dfMoy est correctement dÃ©fini avec les donnÃ©es

# Convertir les colonnes de dates au format datetime
dfMoy['Start date'] = pd.to_datetime(dfMoy['Start date'], format='%d/%m/%Y')
dfMoy['end date'] = pd.to_datetime(dfMoy['end date'], format='%d/%m/%Y')

dfMoy = dfMoy.sort_values(by='Start date')

# Filtrer les donnÃ©es pour la zone spÃ©cifique et la pÃ©riode de temps
zone_data = dfMoy[(dfMoy['zcta'] == 95128) & (dfMoy['Start date'] >= '01/01/2021') & (dfMoy['end date'] <= '31/12/2022')]

# CrÃ©er un graphe
fig, ax1 = plt.subplots(figsize=(12, 6))

# Courbe des cas confirmÃ©s
cas_confirmes_smooth = make_interp_spline(zone_data['Start date'].values.astype(float), zone_data['case count'].values, k=3)
date_smooth = pd.date_range(zone_data['Start date'].min(), zone_data['Start date'].max(), freq='D')
ax1.plot(date_smooth, cas_confirmes_smooth(date_smooth), label='Cas ConfirmÃ©s', color='orange')  # Nouvelle couleur orange

ax1.set_xlabel('Date')
ax1.set_ylabel('Cas ConfirmÃ©s', color='orange')
ax1.tick_params('y', colors='orange')

# CrÃ©er un axe secondaire (partageant le mÃªme axe x)
ax2 = ax1.twinx()
tests_effectues_smooth = make_interp_spline(zone_data['Start date'].values.astype(float), zone_data['test count'].values, k=3)
ax2.plot(date_smooth, tests_effectues_smooth(date_smooth), label='Tests EffectuÃ©s', color='purple')  # Nouvelle couleur pourpre
ax2.set_ylabel('Tests EffectuÃ©s', color='purple')
ax2.tick_params('y', colors='purple')

# CrÃ©er un autre axe secondaire (partageant le mÃªme axe x)
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Ajuster la position du nouvel axe
tests_positifs_smooth = make_interp_spline(zone_data['Start date'].values.astype(float), zone_data['positive tests'].values, k=3)
ax3.plot(date_smooth, tests_positifs_smooth(date_smooth), label='Tests Positifs', color='green')  # Nouvelle couleur verte
ax3.set_ylabel('Tests Positifs', color='green')
ax3.tick_params('y', colors='green')

# Ajouter des labels et un titre
plt.title('Ã‰volution des Cas ConfirmÃ©s, Tests EffectuÃ©s et Tests Positifs (Zone 95128)')

# Afficher le graphe
plt.show()
```
